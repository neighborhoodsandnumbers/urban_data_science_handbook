```python
# Quick demonstration: phase portrait of a simple ODE
t = np.linspace(0, 10, 300)
fig, ax = plt.subplots(figsize=(8, 4))
for omega in [1.0, 1.5, 2.0]:
    ax.plot(t, np.exp(-0.2 * t) * np.cos(omega * t), label=f'$\\omega = {omega}$')
ax.set_xlabel('Time $t$')
ax.set_ylabel('$x(t)$')
ax.set_title('Damped Oscillations')
ax.legend()
plt.tight_layout()
plt.show()
```
