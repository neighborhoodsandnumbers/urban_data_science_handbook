## How big should a city be?
When the negative effects outweigh the positive effects
There's no universal constant here... the constants tend to be at the person level (e.g. Marchetti's Constant, shelling segregation), some of which is cultural (e.g. "american dream")
The max size of a roman or medieval city is much smaller than possible today. Perhaps this can change in the future with more efficient transport, crime management, etc.
