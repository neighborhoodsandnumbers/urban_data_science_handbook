# Urban Economics

## Thoughts on 15-minute city

...it's a decent idea, in theory that all necessary amenities should be within a 15 minute walk. Indeed, let us not have perfection stand in the way of improvement! ... but it is also a misguided and oversimplified concept. As we've learned, spatial proximity is one of the key driving forces behind urbanization. There are faux "constants" in urban science like Marchetti's Constant, which states that people are willing to commute roughly 1 hour in total each day. This is a useful proxy, for sure, but a constant, it is not. And it suggests that people are actually happier to travel more than 15 minutes to some important things daily. Instead of a 15 minute cities, we should think that different things have different travel lengths, which go shrink with density & city size. We might strive for a "2 Minute City" when it comes to reaching a neighbor you trust (15 minutes would be much too far!) or a 1 hour city when it comes to finding a quality sitar repair store (an amazing thing to have in a city, but not on every block). One could even argue that it's the 15+ minute things that shrink their distance fastest in a large city, and ultimately make large cities such unique places. A sitar repair store might be 15 hours away in Iowa, but 45 minutes away in New York. In addition to niche stores, major stadiums and event spaces, job specialization (e.g. medicine) make cities desireable over-and-above 15 minute cities. So, while 15 minutes is nice for many things, let's not forget the 2 minute city (reliable neighbors) and the 90 minute city (sitar repair stores) as well! (NOTE maybe switch NYC with Mumbai here)
